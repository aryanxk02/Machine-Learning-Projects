```python
from matplotlib import pyplot as plt
```

```python
x = [2,1,3,4,5]
y = [10,24,36,40,5]


a = ['india', 'usa', 'china', 'uk', 'new zealand']  #india is on position 2 in bar graph

plt.bar(x, y, tick_label = a, width = 0.5, color = "red")
plt.xlabel('COUNTRIES')
plt.ylabel('POPULAR MONUMENTS')
plt.title('BAR PLOT')
plt.show()
```
