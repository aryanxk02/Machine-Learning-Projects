```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,4,5,7,8,9,11,15,19]
y = [1,4,9,13,20,40,53,69,75,100,225]
```

```python
plt.scatter(x,y, color="red",label = "stars", s =100, marker = "*")

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('SCATTER PLOT')

plt.legend()

plt.show()
```
