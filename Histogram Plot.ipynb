```python
from matplotlib import pyplot as plt
```

```python
marks = [30,28,28,28,29,27,27,27,25,25,20,20,20,20,15,17,18,13]
bins = 40 #number of bars
range = (0,30) #x axis parameters length

plt.hist(marks, bins, range, width = 1.3)
plt.xlabel('marks')
plt.ylabel('number of students')
plt.title('maths exam histogram')
plt.show()
```
